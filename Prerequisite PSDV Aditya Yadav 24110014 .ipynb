Question no: 22 Normalization

```python
data = [2, 2, 5, 4, 2]
normalized=[]
for x in data:
    min_val = min(data)
    max_val = max(data)
    y=(x - min_val) / (max_val - min_val)
    normalized.append(y)
import matplotlib.pyplot as ply
import numpy as np

x=np.array(data)
y=np.array(normalized)
ply.plot(x,y)
ply.show()
```
